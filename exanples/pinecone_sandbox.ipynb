### Query

```python
# Load text embedding vectors already prepated
data = np.loadtxt('data/values.txt', delimiter=',')
values = data.tolist()
```

```python
index = pc.Index(name='general')

query_response = index.query(
    namespace="",
    vector=values,
    top_k=5,
    include_values=True,
    include_metadata=True
)
```
